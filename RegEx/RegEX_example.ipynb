```python
import re
```

```python
text="This Course is developed by USMC Central Command. The dates are 15.04.2027 to 12.06.2028 15/May/2027 to 12/June/2028. 15.May.2027 to 12.June.2028 Classes 3-7 All submission should be sent to major_sam_jackson@usmc.mil.test \
      Stay alert until further notice. Notice Code D23Xq456 /xq1/124"
```

```python
pattern=r"\d{2}[./](\d{2}|[a-zA-Z]{3,})[./]\d{4}"
#pattern=r"\d{2}[./](d{2,}|[a-zA-Z]{2,})\d{4}"
```

```python
matches=re.finditer(pattern,text)
```

```python
for match in matches:
    print(match.group())
```
