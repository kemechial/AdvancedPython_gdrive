```python
import mechanicalsoup
import html5lib
browser = mechanicalsoup.StatefulBrowser()
```

```python
 browser.open("https://webscraper.io/test-sites/e-commerce/more")
```

```python
browser.url
```

```python
page=browser.request('GET','https://webscraper.io/test-sites/e-commerce/more').soup
```

```python
print(page.prettify())
```

```python
products=page.find_all('div',{'class':'col-md-4 col-xl-4 col-lg-4'})
```

```python
print(products)
```

```python
with open("/gdrive/My Drive/İleriPython/WebScraping/products.csv","w",encoding="utf-8") as file:

    csv_writer=writer(file)
    csv_writer.writerow(["product","specifications"])

    for product in products:
     title=product.find("a",{'class':'title'}).get_text()
     spec=product.find("p",{'class':'description'}).get_text()
     csv_writer.writerow([title,spec])
```
