```python
def counter1(max):
  x=1
  while x <= max:
    yield x
    x+=1
```

```python
gen1=counter1(10)
```

```python
print(next(gen1))
```

```python
print(next(gen1))
```

```python
for i in gen1:
  print(i)
```
