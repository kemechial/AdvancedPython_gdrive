```python
class Movie:
    def __init__(self, title, director, year,duration):
        self.title = title
        self.director = director
        self.year = year
        self.duration = duration

    def __repr__(self):
        return f"The movie {self.title} was produced in {self.year} and directed by {self.director}"

    def __len__(self):
        return self.duration

    def __del__(self):
        print(f"The movie {self.title} was deleted")
```

```python
m1=Movie('Alien','Ridley Scott',1979, 138)
```

```python
print(m1)
```

```python
len(m1)
```

```python
del m1
```
