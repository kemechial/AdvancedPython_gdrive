```python
class Human:
  def __init__(self, name, surname, age):
    self.name = name
    self.surname = surname
    self.age = age
    print("New human Instance")

  def intro(self):
    print(f"Hello my name is {self.name} {self.surname}. I am  {self.age} years old.")


class Student(Human):
  def __init__(self, name, surname, age, student_id):
    super().__init__(name, surname, age)
    self.student_id = student_id
    print("New Student Instance")

  def study(self):
    print(f"Student {self.name} is studying.")

  def intro(self):
    super().intro()
    print(f"Hello my name is {self.name} {self.surname}. I am  {self.age} years old. My number is {self.student_id}")

class Teacher(Human):
  def __init__(self, name, surname, age, branch):
    Human.__init__(self, name, surname, age)
    self.branch = branch
    #super().__init__(name, surname, age)
    print("New Student Instance")

  def teach(self):
    print(f"Teacher {self.name} is teaching {self.branch}.")
```

```python
t1=Teacher("John", "Doe", 30, "physics")
```
