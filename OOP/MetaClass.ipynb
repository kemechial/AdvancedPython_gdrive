```python
class Meta(type):
  def __new__(self, class_name, bases, attrs):
    print(attrs)

    a={}
    for k,v in attrs.items():
      if k.startswith("_"):
        a[k]=v
      else:
        a[k.upper()]=v

    return type(class_name, bases, attrs)
```

```python
class Person(metaclass=Meta):
  x=5
  y=10
  def hello(self):
    return "hello"
```

```python
p=Person()
```
