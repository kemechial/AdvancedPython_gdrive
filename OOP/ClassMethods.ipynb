```python
class ShopItem:
    discount_rate = 0.1 #class attribute
    item_count=0

    @classmethod
    def display_item_count(cls):
        print(f"There are {cls.item_count} items in the shop.")

    @classmethod
    def create_item(cls, str):
        name, price, quantity = str.split(",")
        return cls(name, price, quantity)

    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price #instance attribute
        self.quantity = quantity
        ShopItem.item_count+=1

    def __str__(self):
        return f"{self.name} - ${self.price:.2f}"

    def calculate_total(self):
        return self.price * self.quantity

    def discount(self):
      self.price = self.price * (1-ShopItem.discount_rate)
```

```python
item1=ShopItem("Shirt", 19.99, 3)
item2=ShopItem("Jeans", 49.99, 4)
item3=ShopItem("Shoe", 29.99, 2)
item4=ShopItem("Test", 49.99, 4)
```

```python
ShopItem.display_item_count()
```

```python
ShopItem.create_item("Shirt,19.99,3")
```

```python
ShopItem.display_item_count()
```

```python
class Coupon:
  def __init__(self, code, discount):
    self.code=code
    self.discount=discount
```

```python
coupon1=Coupon("SAVE10", 0.2)
coupon2=Coupon("FREESHIP", 0.3)
coupon3=Coupon("NEWUSER", 0.1)
```

```python
class ShoppingCart:
  coupon_list=[coupon1,coupon2,coupon3]

  def __init__(self,itemList):
    self.itemList=itemList

  def add_item(self, item):
    self.itemList.append(item)

  def display_items(self):
    for item in self.itemList:
      print(f"{item.name} is {item.price}")

  def calculate_total(self):
    for item in self.itemList:
      total=sum([item.calculate_total() for item in self.itemList])
    return total

  def remove_item(self, cart_item):
    self.itemList=[item for item in self.itemList if item!=cart_item]

  @classmethod
  def get_coupons(cls):
    return [c.code for c in cls.coupon_list]

  @classmethod
  def get_coupon(cls, code):
   # return [coupon for coupon in cls.coupon_list if coupon.code==code]
   return next(filter(lambda c: c.code==code,cls.coupon_list)) #return first found item

  def apply_coupon(self, code):
    if code not in ShoppingCart.get_coupons():
      raise ValueError(f"invalid code: {code}")
    coupon=ShoppingCart.get_coupon(code)

    for item in self.itemList:
      item.price=item.price*(1-coupon.discount)
```

```python
sc=ShoppingCart([item1,item2,item3,item4])
sc.display_items()
```
