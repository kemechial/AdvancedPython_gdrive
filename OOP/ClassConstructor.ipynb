```python
class ShopItem:
    discount_rate = 0.1 #class attribute
    item_count=0
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price #instance attribute
        self.quantity = quantity
        ShopItem.item_count+=1

    def __str__(self):
        return f"{self.name} - ${self.price:.2f}"

    def calculate_total(self):
        return self.price * self.quantity

    def discount(self):
      self.price = self.price * (1-ShopItem.discount_rate)
```

```python
item1=ShopItem("Shirt", 19.99, 3)
item2=ShopItem("Jeans", 49.99, 4)
```
