```python
class MyNumbers:
  def __init__(self,start,stop):
    self.start=start
    self.stop=stop

  def __iter__(self):
    return self

  def __next__(self):
    if self.start<=self.stop:
      x=self.start
     # self.start+=1
      self.start+=3
      return x
    else:
      raise StopIteration
```

```python
iterator_list=iter(MyNumbers(1,10))
```

```python
for i in iterator_list:
  print(i)
```

```python
#iterator depleted above define again
iterator_list=iter(MyNumbers(1,10))
while True:
  try:
    element=next(iterator_list)
    print(element)
  except StopIteration:
    break
```
