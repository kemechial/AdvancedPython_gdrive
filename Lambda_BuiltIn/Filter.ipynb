```python
numbers=[x for x in range(0,101) if x%2==0 and x%3==1]
numbers
```

```python
filtered=list(filter(lambda x:x%4==0,numbers))
filtered
```

```python
def filter_func(x):
    if x%5==0:
      return True
    else:
      return False
```

```python
filtered2=list(filter(filter_func,numbers))
filtered2
```

```python
people=[{"name":"ali", "grades":[0,10,20]},{"name":"ayşe", "grades":[30,0,10,20]},{"name":"jack", "grades":[0,10,20,40,60]}]
```

```python
filtered_people=list(filter(lambda x: x["name"]=="ali",people))
filtered_people
```

```python
filtered_people=list(filter(lambda x: len(x["grades"])>3, people))
filtered_people
```

```python
users=list(map(lambda x: x["name"], people))
users
```

```python
result=[x["name"].upper() for x in people if len(x["grades"])>1]
result
```
