```python
students=['ram','shyam','hari','gita','sita','geeta','john','alex']
grades=[10,100,45,78,90,5,0,57]
```

```python
combined=[(students[s],grades[s]) for s in range(0,len(students))]
combined
```

```python
best_grades={key:value for (key,value) in combined if value>50}
best_grades
```
