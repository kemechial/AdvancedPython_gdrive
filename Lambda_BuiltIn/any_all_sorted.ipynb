```python
test_list=[{"name":"jack","notes":[3,4,6,8]},{"name":"joe","notes":[3,6,7]},{"name":"edward","notes":[3,4,6,8,14]}]
```

```python
sorted_people=sorted(test_list,key=lambda user: len(user["notes"]), reverse=True)
print(sorted_people)
```
