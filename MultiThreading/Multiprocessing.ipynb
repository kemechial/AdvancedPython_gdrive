```python
import time
import multiprocessing
list_square=[]
list_cube=[]


def square(num_list,list1):

  for i,n in enumerate(numList):
    time.sleep(0.1)
    print("square",n*n)
    list1[i]=n*n

def cube(num_list,list1):

  for i,n in enumerate(numList):
    time.sleep(0.3)
    print("cube",n*n*n)
    list1[i]=n*n*n

if __name__ == "__main__":
   #define in shared memory
   list_square=multiprocessing.Array("i",10)
   list_cube=multiprocessing.Array("i",10)

   t=time.time()

   p1=multiprocessing.Process(target=square,args=(numList,list_square))
   p2=multiprocessing.Process(target=cube,args=(numList,list_cube))

   p1.start()
   p2.start()
   p1.join()
   p2.join()
   print("job finished...(Multiprocessing)",time.time()-t)
   print(list_square[:])
   print(list_cube[:])
```
