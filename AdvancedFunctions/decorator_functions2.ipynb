```python
import time
```

```python
def speed_test(fn):
  def inner(*args, **kwargs):
    start_time = time.perf_counter()
    print(f"fn.__name__ fonksiyonu calisiyor.")
    result = fn(*args, **kwargs)
    end_time = time.perf_counter()
    run_time = end_time - start_time
    print(f"Function {fn.__name__} took {end_time - start_time} seconds to run.")
    return result
  return inner


@speed_test
def sum_gen():
    return sum((x for x in range(int(1e6))))

@speed_test
def sum_list():
    return sum([x for x in range(int(1e6))])
```

```python
sum_gen()
sum_list()
```
