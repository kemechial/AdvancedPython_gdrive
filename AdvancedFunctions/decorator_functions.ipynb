```python
def greeting(fn):
 def inner(name):
  print("Good morning")
  fn(name)
  print("see you")
 return inner


@greeting
def welcome(name):
  print(f"welcome to python Mr. {name}")


@greeting
def ask(name):
  print(f"What do you drink Mr. {name}?")
```

```python
welcome("John")
```

```python
ask("John")
```

```python
def double_fun(fn):
    def inner(*args,**kwargs):
     fn(*args,**kwargs)
     fn(*args,**kwargs)
    return inner
```

```python
@double_fun
def welcome():
    print("welcome to python")
```

```python
@double_fun
def ask(name):
    print("What do you drink?",name)
```

```python
def double_fun_r(fn):
    def inner(*args,**kwargs):
     return fn(*args,**kwargs)
    return inner
```

```python
@double_fun_r
def ask(name):
    return "What do you drink?"
```

```python
welcome()
```
