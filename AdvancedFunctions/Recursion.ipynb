```python
def factorial(num):
  if not isinstance(num,int):
    raise TypeError("num must be a positive int")
  if not num>=0:
    raise ValueError("num must be zero or positive")
  def inner_factorial(num):
    if num<=1:
      return 1
    return num*inner_factorial(num-1)
  return inner_factorial(num)
```

```python
factorial(5)
```

```python
try:
  factorial(-5)
except Exception as e:
  print(e)
```
