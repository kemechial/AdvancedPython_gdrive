```python
 def dec_greeting(count):
    def greeting(fn):
      def inner(name):
        for _ in range(count):
         fn(name)
      return inner
    return greeting

@dec_greeting(count=2)
def welcome(name):
  print(f"welcome to python Mr. {name}")


@dec_greeting(count=2)
def ask(name):
  print(f"What do you drink Mr. {name}?")
```

```python
welcome("John")
ask("Jane")
```
