```python
def numPower(base):
  def inner(power):
    return base ** power
  return inner

numPowerOfTwo = numPower(2)(4)
print(numPowerOfTwo)

numPowerOfThree = numPower(3)(4)
print(numPowerOfThree)
```

```python
def yetkiSorgula(page):
  def inner(role):
    if role == 'Admin':
      return "{0} rolü {1} sayfasına ulaşabilir.".format(role, page)
    else:
      return "{0} rolü {1} sayfasına ulaşamaz.".format(role, page)
  return inner
```

```python
yetkiSorgula('Product Edit')('Admin')
```

```python
def islem(islemAdi):
  def toplam(*args):
    toplam = 0
    for arg in args:
      toplam += arg
    return toplam
  def carpma(*args):
    carpim = 1
    for arg in args:
      carpim *= arg
    return carpim
  if islemAdi == 'toplama':
    return toplam
  else:
    return carpma
```

```python
toplama = islem('toplama')
print(toplama(1, 2, 3, 4, 5, 6))
#
```

```python
carpma = islem('carpma')
print(carpma(1, 2, 3, 4, 5, 6))
#
```
